Handling state territory & postalcode with "Unknown"

```python

filtered_data['STATE'] = filtered_data['STATE'].fillna('Unknown')  # Fill missing 'State' values with 'Unknown'
filtered_data['TERRITORY'] = filtered_data['TERRITORY'].fillna('Unknown')
filtered_data['POSTALCODE'] = filtered_data['POSTALCODE'].fillna('Unknown')
```

# Correcting inconsistent formatting

Converting all the string data into lowercase

- .str.lower():all string in a single column to lowercase
- .apply():allows us to apply a function across the rows or columns of DataFrame
- .select_dtypes():allows to select columns of a specific data type in the DataFrame
    - used to filter columns based on their data types
- The lambda function:a reserved keyword to define anonymous
    - lambda arguments: expression
    - is used to apply a transformation to each column of the DataFrame.
    - it checks if the columns data type is "object" 
        - then replaces with lowercase if true else return x

```python
managed_data=managed_data.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
managed_data.head()
```

Removing extra spaces
- apply() vs applymap()
    - apply() is used applying a function along an axis (rows or columns) of the DataFrame.
    - applymap() is used when applying a function element-wise to every single value in DataFrame.
- applymap(): Applies a function element-wise to all values in the DataFrame.
- isinstance(x, str):check the type of an object x. 
    - Here, it checks whether the object x is an instance of the str class
- when using applymap future warning arrived


```python
managed_data= managed_data.apply(lambda x: x.strip() if isinstance(x, str) else x)
managed_data.head()
```
