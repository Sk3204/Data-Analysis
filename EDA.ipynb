# Analyzing the total sales and profit% and total qty sold of each columns

grouping same type of productline rows and calculating

trnsformed_data['PRODUCTLINE'] == 'Motorcycles'
- This creates a boolean mask (True/False) indicating which rows have 'PRODUCTLINE' as 'Motorcycles'.

trnsformed_data[...]
- filters rows where the condition is True.
- creates new DataFrame that only contains rows with required true data


Grouping productline and calculating

.agg():ggregates data, allowing you to apply multiple functions to a DataFrame at once
- result = df.agg({'Sales': ['sum', 'mean', 'max', 'min']})

.df.groupby('Category') groups data by 'Category'

.assign() adds multiple new columns in a single chain, keeping code concise.

Performs only for motorcycle

```python
mgrouped_data= (transformed_data[transformed_data['PRODUCTLINE'] == 'motorcycles']
        .groupby('PRODUCTLINE')
        .agg(
            T_sales=('SALES', 'sum'),              # Total sales
            T_qty=('QTY', 'sum'),                  # Total quantity
            T_cost=('COST', 'sum'),                 # Total cost
            T_profit=("PROFIT","sum")

        )
        .assign(
            T_profit_pct=lambda g_data:round((g_data["T_profit"]/g_data["T_sales"])*100,2)
        )
)
mgrouped_data
```

Performs for all Productline at once

```python
productline_data=( 
    transformed_data.groupby('PRODUCTLINE')#groups data by pl so each pl gets aggregated separately
    .agg(
        T_sales=('SALES', 'sum'),              # Total sales
        T_qty=('QTY', 'sum'),                  # Total quantity
        T_cost=('COST', 'sum'),                # Total cost
    )#computes total sales qty cost for each productline(pl)
    .assign(
        T_profit=lambda g_data:g_data["T_sales"]-g_data["T_cost"],
        T_profit_pct=lambda g_data:round((g_data["T_profit"]/g_data["T_sales"])*100,2)
    )
)
productline_data
```

# Analyzing all by using single Analyzing function

```python
def Analyze(field):
    grouped_data=(
        transformed_data.groupby(field)#groups data by pl so each pl gets aggregated separately
        .agg(
        T_sales=('SALES', 'sum'),              # Total sales
        T_qty=('QTY', 'sum'),                  # Total quantity
        T_cost=("COST","sum"),
        T_profit=("PROFIT","sum")
        )
        .assign(
        T_profit_pct=lambda g_data:round((g_data["T_profit"]/g_data["T_sales"])*100,2)
        )
    )
    return grouped_data
```

```python
transformed_data.head()
```

```python
product_data=Analyze("PRODUCTLINE")
country_data=Analyze("COUNTRY")
city_data=Analyze("CITY")
state_data=Analyze("STATE")
territory_data=Analyze("TERRITORY")
dealsize_data=Analyze("DEALSIZE")
status_data=Analyze("STATUS")

yearly_data=Analyze(transformed_data["DATE"].dt.year)
monthly_data=Analyze(transformed_data["DATE"].dt.month)
day_data=Analyze(transformed_data["DATE"].dt.day)
```
